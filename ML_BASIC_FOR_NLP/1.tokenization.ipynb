## Tokenization

```python
corpus="""Hello, I am Abhipsa. I am a student in my final year of college.
I am learning GENAI. My aim to become AI engineer and build system that bring real revolutions. Wish me luck!"""
print(corpus)
```

## Paragraphs to sentences

```python
from nltk.tokenize import sent_tokenize
docs=sent_tokenize(corpus)
```

```python
for i in docs:
    print(sent_tokenize(i))
```

## Sentences into words
- Each and every word has different meaning in the sentence

```python
from nltk.tokenize import word_tokenize
word_tokenize(corpus)
```

```python
for sentence in docs:
    print(word_tokenize(sentence))
```

```python
from nltk.tokenize import wordpunct_tokenize

wordpunct_tokenize(corpus)
```

```python
from nltk.tokenize import TreebankWordTokenizer
tokenizer=TreebankWordTokenizer()
```

```python
tokenizer.tokenize(corpus)
```
