## Stemming
- Stemming is the process of reducing a word to its word stem that affixes to suffixes and prefixes or to the roots of words known as a lemma.
- Stemming is important in natural language understanding(NLU) and natural language processing(NLP)

#### Classification problem
- comments of the product is a positive review or negative review
- reviews --> [eating, eat, eaten]-->eat,[going,goes,gone]->go

```python
words = ["eating","eat","eaten","writing","writes","written","programming","programs","programmer","going","goes","history","finally"]
```

## PorterStemmer

```python
from nltk.stem import PorterStemmer
stemming = PorterStemmer()
```

```python
for word in words:
    print(word +"------->"+stemming.stem(word))
```

```python
stemming.stem('congratulations')
```

- With some words it works good but with some words it generates issues

### RegexpStemmer class
NLTK has RegexpStemmer class with the help of which we can easily implement regular expression stemmer algorithms. It basically takes a single regualr expression and removes any prefix or suffix that matches the eexpression.

```python
from nltk.stem import RegexpStemmer
reg_stemmer=RegexpStemmer('ing$|s$|e$|able$',min=4)
reg_stemmer.stem('eating'), reg_stemmer.stem('igneating')
```

## SnowBall Stemmer
- better than porterstemmer

```python
from nltk.stem import SnowballStemmer
sn=SnowballStemmer(language='english')
```

```python
for word in words:
    print(word+"-->"+sn.stem(word))
```

```python
stemming.stem('fairly')
```
